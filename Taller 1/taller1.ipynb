#### iv

```python
#Vamos a crear la columna Elite

college['Elite']=['No' for x in range(len(college)) ]
```

```python
#Ahora reemplazamos los No por Yes si se cumple Top10perc>50
for x in range(len(college)):
    if college['Top10perc'][x]>50:
        college.set_value(x, 'Elite', 'Yes')
```

```python
#Vemos cuantas universidades son de la élite
college['Elite'].describe()
```

```python
#Seleccionamos los datos a pintar
si_e=college[college['Elite']=='Yes']
no_e=college[college['Elite']=='No']
```

```python
#Pintamos
trace0 = go.Box(
    y=si_e['Outstate'],
    name = 'Elite',
    marker = dict(
        color = 'rgb(214, 12, 140)',
    )
)
trace1 = go.Box(
    y=no_e['Outstate'],
    name = 'No Elite',
    marker = dict(
        color = 'rgb(0, 128, 128)',
    )
)
layout = go.Layout(
    title = "Boxplots Outstate y Elite"
)
data = [trace0, trace1]
fig = go.Figure(data=data,layout=layout)
py.iplot(data)
```

### Conclusión 2

Para estudiantes de afuera del estado, estudiar en una universidad cara (del tercer cuartil hacia arriba) de la élite no vale mucho más que una cara que no sea de la élite.

Vamos a atacar cuatro situaciones:

```python
#1.
#Primero seleccionemos las universidades públicas
publicas=college[college['Private']=='No']
```

```python
publicas.head()
```

```python
#Ahora cojamos los features que necesitamos
publicas_temp=publicas[['Outstate', 'Elite']]
```

```python
#Finalmente los outliers...
outliers_publicas=publicas_temp[publicas_temp['Outstate']>np.percentile(publicas_temp['Outstate'],q=75)]
```

Veamos si quizá son más caras porque son de la élite...

```python
#Importemos el paquete para contar elementos
from scipy.stats import itemfreq
```

```python
itemfreq(outliers_publicas['Elite'])
```

```python
itemfreq(publicas['Elite'])
```

### Conclusión 3

Más de la mitad de los outliers son de la élite

```python
#Ataquemos ahora el item 2.
college['perc accept']=100*college['Accept']/college['Apps']
```

```python
publicas=college[college['Private']=='No']
privadas=college[college['Private']=='Yes']
```

```python
#Analicemos por boxplots...

trace0 = go.Box(
    y=privadas['perc accept'],
    name = 'Privadas',
    marker = dict(
        color = 'rgb(0, 128, 128)',
    )
)
trace1 = go.Box(
    y=publicas['perc accept'],
    name = 'Publicas',
    marker = dict(
        color = 'rgb(214, 12, 140)',
    )
)
layout = go.Layout(
    title = "Boxplots tasa aceptación privadas y públicas"
)
data = [trace0, trace1]
fig = go.Figure(data=data,layout=layout)
py.iplot(data)
```

```python
itemfreq(privadas[privadas['perc accept']<np.percentile(privadas['perc accept'], q=25)]['Elite'])
```

```python
itemfreq(privadas['Elite'])
```

### Conclusión 4

Entrar a las universidades de eĺite privadas es significativamente más difícil que entrar a las públicas.

Los estudiantes top prefieren ir a universidades privadas que públicas

```python
#Para terminar hagamos el 4.
#Analicemos por boxplots...

trace0 = go.Box(
    y=privadas['Grad.Rate'],
    name = 'Privadas',
    marker = dict(
        color = 'rgb(0, 128, 128)',
    )
)
trace1 = go.Box(
    y=publicas['Grad.Rate'],
    name = 'Publicas',
    marker = dict(
        color = 'rgb(214, 12, 140)',
    )
)
layout = go.Layout(
    title = "Boxplots tasa de graduación privadas y públicas"
)
data = [trace0, trace1]
fig = go.Figure(data=data,layout=layout)
py.iplot(data)
```
