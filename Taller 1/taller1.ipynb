```python
#Calculamos aquí los coeficientes de correlación para hablar con cifras.
import scipy.stats
scipy.stats.spearmanr(cuanti[['weight', 'horsepower', 'displacement', 'mpg', 'acceleration']])
```

Hay una relación inversa muy similar entre mpg vs. displacement,mpg vs horsepower y mpg vs. weight. Esta semejanza se explica en la casi que perfecta relación entre weight y displacement (un factor de 0.93) y la alta entre horsepower y estas dos.

Entre el resto de variables no se envidencia relación fuerte.

```python
#Ahora veamos las discretas. Convertiremos todas las variables discretas
#a 3 o menos categorias para visualizar mejor. Para ello definamos
#algunas funciones auxiliares.
def convertidor_hp(x):
    if x <= np.percentile(cuanti['horsepower'], q=33):
        return 0
    elif (x>np.percentile(cuanti['horsepower'], q=33)) & (x <= np.percentile(cuanti['horsepower'], q=66)):
        return 1
    else:
        return 2

def convertidor_year(x):
    if x <= np.percentile(cuanti['year'], q=33):
        return 0
    elif (x>np.percentile(cuanti['year'], q=33)) & (x <= np.percentile(cuanti['year'], q=66)):
        return 1
    else:
        return 2
        
def convertidor_cylinder(x):
    if x <= np.percentile(cuanti['cylinders'], q=33):
        return 0
    elif (x>np.percentile(cuanti['cylinders'], q=33)) & (x <= np.percentile(cuanti['cylinders'], q=66)):
        return 1
    else:
        return 2
        
        
```

```python
#APlicamos las funciones a las tres variables
cuanti['cat_hp']=[convertidor_hp(x) for x in cuanti['horsepower']]
cuanti['cat_year']=[convertidor_year(x) for x in cuanti['year']]
cuanti['cat_cylinder']=[convertidor_cylinder(x) for x in cuanti['cylinders']]

#DEfinamos los subsets que vamos a pintar
hp1=cuanti[cuanti['cat_hp']==0]
hp2=cuanti[cuanti['cat_hp']==1]
hp3=cuanti[cuanti['cat_hp']==2]

year1=cuanti[cuanti['cat_year']==0]
year2=cuanti[cuanti['cat_year']==1]
year3=cuanti[cuanti['cat_year']==2]

cy1=cuanti[cuanti['cat_cylinder']==0]
cy2=cuanti[cuanti['cat_cylinder']==1]
cy3=cuanti[cuanti['cat_cylinder']==2]
```
