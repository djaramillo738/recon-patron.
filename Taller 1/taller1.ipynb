#### iv

```python
#Vamos a crear la columna Elite

college['Elite']=['No' for x in range(len(college)) ]
```

```python
#Ahora reemplazamos los No por Yes si se cumple Top10perc>50
for x in range(len(college)):
    if college['Top10perc'][x]>50:
        college.set_value(x, 'Elite', 'Yes')
```

```python
#Vemos cuantas universidades son de la lite
college['Elite'].describe()
```

```python
#Seleccionamos los datos a pintar
si_e=college[college['Elite']=='Yes']
no_e=college[college['Elite']=='No']
```

```python
#Pintamos
trace0 = go.Box(
    y=si_e['Outstate'],
    name = 'Elite',
    marker = dict(
        color = 'rgb(214, 12, 140)',
    )
)
trace1 = go.Box(
    y=no_e['Outstate'],
    name = 'No Elite',
    marker = dict(
        color = 'rgb(0, 128, 128)',
    )
)
layout = go.Layout(
    title = "Boxplots Outstate y Elite"
)
data = [trace0, trace1]
fig = go.Figure(data=data,layout=layout)
py.iplot(data)
```

### Conclusin 2

Para estudiantes de afuera del estado, estudiar en una universidad cara (del tercer cuartil hacia arriba) de la lite no vale mucho ms que una cara que no sea de la lite.

```python
publicas.head()
```

```python
#Ahora cojamos los features que necesitamos
publicas_temp=publicas[['Outstate', 'Elite']]
```

Veamos si quiz son ms caras porque son de la lite...

```python
itemfreq(outliers_publicas['Elite'])
```

```python
publicas=college[college['Private']=='No']
privadas=college[college['Private']=='Yes']
```

```python
#Analicemos por boxplots...

trace0 = go.Box(
    y=privadas['perc accept'],
    name = 'Privadas',
    marker = dict(
        color = 'rgb(0, 128, 128)',
    )
)
trace1 = go.Box(
    y=publicas['perc accept'],
    name = 'Publicas',
    marker = dict(
        color = 'rgb(214, 12, 140)',
    )
)
layout = go.Layout(
    title = "Boxplots tasa aceptacin privadas y pblicas"
)
data = [trace0, trace1]
fig = go.Figure(data=data,layout=layout)
py.iplot(data)
```
