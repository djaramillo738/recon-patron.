# Capítulo 7 ejercicio 6 

```python
Wage=pd.read_csv("data/Wage.csv")
```

```python
Wage.head()
```

```python
age=Wage["age"]
wage=Wage["wage"]
plt.scatter(Wage["age"],Wage["wage"], alpha=0.5, c='#4a1f3d')
plt.xlabel('age')
plt.ylabel('wage')
plt.show()
```

```python
coef=np.polyfit(age, wage,4)
p=np.poly1d(coef)
plt.xlabel("age")
plt.ylabel("wage")
plt.plot(age[1],p(age[1]), c='#00a9ff', label='cuártica')
#plt.ylim(40,300)
x_new = np.linspace(age[0], age[2999], 2980)
y_new = p(x_new)

plt.plot(age,wage,'o', x_new, y_new)
plt.xlim(10, 90)
plt.show()
```
