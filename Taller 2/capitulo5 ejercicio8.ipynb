```python
#Hacemos la regresión polinomial para todos los grados y clculamos su respectivo error cuadrático
def reg_pol(n):
    dis=df["dis"]
    nox=df["nox"]
    coef=np.polyfit(dis, nox,n)
    p=np.poly1d(coef)
    plt.xlabel("dis")
    plt.ylabel("nox")

    x_new = np.linspace(dis[0], age[505], 500)
    y_new = p(x_new)

    plt.plot(dis,nox,'o', x_new, y_new)
    plt.xlim(0, 12)
    plt.ylim(0.3,1)
    plt.show()
    print(np.sum((np.polyval(np.polyfit(dis, nox, n), dis) - nox)**2))
```

```python
[reg_pol(i) for i in range(9)]
```

```python
#El error cuadrático va disminuyendo notablemente hasta llegar a grado 3, de ahí en adelante disminuye muy levemente. Por lo que el mejor grado para hacer esta regresión es n=3
```

```python
from patsy import dmatrix
import statsmodels.api as sm
from sklearn import preprocessing
import statsmodels.api as sm
import statsmodels.formula.api as smf
dis_grid = np.arange(df.dis.min(), df.dis.max()).reshape(-1,1)
```

```python
transformed_x3 = dmatrix("cr(df.age, df=4)", {"df.age": df.age}, return_type='dataframe')
fit3 = sm.GLM(df.nox, transformed_x3).fit()
fit3.params

pred3 = fit3.predict(dmatrix("cr(age_grid, df=4)", {"age_grid": age_grid}, return_type='dataframe'))

plt.scatter(df.age, df.nox, facecolor='None', edgecolor='k', alpha=0.1)
plt.plot(age_grid, pred3, color='b', label='Spline con df=4')
plt.legend()
plt.xlim(15,85)
plt.ylim(0,1)
plt.xlabel('dis')
plt.ylabel('nox')

plt.show()
```

```python
def spline_con_df(n):
    transformed_x3 = dmatrix("cr(df.age, df=n)", {"df.age": df.age}, return_type='dataframe')
    fit = sm.GLM(df.nox, transformed_x3).fit()
    fit.params

    pred = fit.predict(dmatrix("cr(age_grid, df=n)", {"age_grid": age_grid}, return_type='dataframe'))

    plt.scatter(df.age, df.nox, facecolor='None', edgecolor='k', alpha=0.1)
    plt.plot(age_grid, pred, color='b', label='Spline con df=n')
    plt.legend()
    plt.xlim(15,85)
    plt.ylim(0,1)
    plt.xlabel('dis')
    plt.ylabel('nox')
    print(fit.params)
    plt.show()

```
